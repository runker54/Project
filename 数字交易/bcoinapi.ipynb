```python
import time
import pandas as pd
import requests
```

```python
BASE_URL = "https://api.binance.com"
url = BASE_URL + "/api/v1/klines?symbol=DOGEUSDT&interval=1h&limit=100"
resp = requests.get(url)
resp = resp.json()
#将时间戳转换为标准时间格式
def timestamp_to_fomat(timestamp=None,format='%Y-%m-%d %H:%M:%S'):
     #默认返回当前格式化好的时间
     #传入时间戳的话，把时间戳转换成格式化好的时间，返回
    if timestamp:
        time_tuple = time.localtime(timestamp)
        res = time.strftime(format,time_tuple)
    else:
         res = time.strftime(format)#默认读取当前时间
    return res
```

```python
df = pd.DataFrame(resp)
df = df.drop(columns=[6,7,8,9,10,11])
df.columns = ["时间","open","high","low","close","总量"]
df["时间"] = (df["时间"]//1000).map(timestamp_to_fomat)
df.set_index(["时间"],drop=True)
```

```python
import cufflinks as cf
from cufflinks import iplot
cf.set_config_file(theme='pearl',sharing='public',offline=True)
df.iplot(kind="ohlc",keys=["open", "high", "low", "colse"],rangeslider=True)
```

```python
qf=cf.QuantFig(df,title='DOGE',legend='top',name='GS')
qf.add_bollinger_bands()
qf.iplot()
```
